```python
importer  des pandas  en tant que  pd

df = pd.read_excel ( '/content/drive/MyDrive/CleanData_par_date.xlsx' )

# Convertir la colonne date en datetime
df [ "date_de_repérage" ]  = pd.to_datetime ( df [ "date_de_repérage" ],  erreurs= 'coerce' )

# Extraire l'année
df [ "année" ]  = df [ "date_de_repérage" ] .dt.year

# Calculer les pourcentages par année
pourcentages = df [ "année" ] .value_counts ( normalize= True )  *  100

# Afficher les pourcentages triés du plus élevé au  plus bas
print ( pourcentages.sort_values ​​( ascending= False ) . round ( 2 ))
```

```python
importateur matplotlib.pyplot en tant que plt

# Réutiliser les pourcentages déjà calculés
pourcentages.sort_values ​​​​()  .plot  (  kind=  'barh' ,   color=  'orange' )

plt.title  ( "Pourcentage de fausses nouvelles par année" )
plt.xlabel  ( "Pourcentage (%)" )
plt.ylabel  ( "Année" )
plt.show  ()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_excel('/content/drive/MyDrive/CleanData_par_thématique.xlsx')

# Filtrer les thématiques souhaitées
thématiques_cibles = ["Santé", "Immigration", "Politique", "Économie"]
df_filtré = df[df["thématique"].isin(thématiques_cibles)]

# Compter et tracer
df_filtré["thématique"].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Thématiques ciblées")
plt.show()
```
